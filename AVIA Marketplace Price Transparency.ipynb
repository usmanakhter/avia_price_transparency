```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("white")
sns.set()
```

# Price Transparency
The AVIA Marketplace lacks price transparency. Below are the prices on all items in the Marketplace.

[AVIA Marketplace Price Transparency Spreadsheet](https://drive.google.com/open?id=198OluXOtVzZ6Icd6EiI54ygvMnXWUbyU-cTiiqFfau4)

```python
df=pd.read_csv("avia_marketplace.csv")
```

```python
print(df.to_string())
```

```python
df_food=df[:162]
df_drinks=df[163:]

#piechart
labels = ['Food', 'Drinks']
sizes = [len(df_food), len(df_drinks)]
colors = ['green', 'orange']
explode = (0.1, 0.1)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

### What are the cheapest items?

```python
sns.catplot(y="price", kind="swarm", color="green", data=df_food);
sns.catplot(y="price", kind="swarm", color="orange", data=df_drinks);
```

The largest cluster of food prices are at \$1. That's also one of the lowest food prices available, with just a handful of items under \$1. Drinks are less clustered. Most drinks fall under \$3. 

```python
df_food.query('price <= 1').sort_values('price')
```

It shouldn't be surprising that the most processed snacks are the cheapest. Fortunately, apples and clementines are even cheaper! And what's that item way up there at \$5.00? Creminelli Casalingo Salami Aged Gouda and Dried Cherries.

```python
df_drinks.query('price <= 2').sort_values('price')
```

Unsurprisingly, soda (pop?) cans are the cheapest drink, with La Croix also at \$1. And what's that drink up there at $5.30? Health-Ade Kombucha Ginger-Lemon.
